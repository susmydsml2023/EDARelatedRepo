<h5><b>4.Find out under which team and position, spending in terms of salary is high.</b></h5>

```python
df["Salary"]
```

```python
positions = df["Position"].unique()

for position in positions:
    positionSalaryList = df[df["Position"] == position][["Name","Position","Salary"]]
    print(f"POSITION '{position}':")
    print(positionSalaryList.to_string(justify='center', index=False))
    total_salary = positionSalaryList["Salary"].sum()
    print(f"Total Salary of '{position}': {total_salary}")
    print("\n")
```

```python
#box plot
sns.boxplot(x='Position',y='Salary',data=df)
```

```python
teams = df["Team"].unique()

for team in teams:
    teamList = df[df["Team"] == team][["Name", "Team", "Salary"]]
    print(f"Team '{team}':")
    total_salary = teamList["Salary"].sum()
    print(f"Total Salary of '{team}': {total_salary}")
    print("\n")
```

```python
plt.figure(figsize=(12, 6))
plt.xticks(rotation=90)
sns.set_theme(style="whitegrid")
sns.barplot(x='Team', y='Salary', data=df, estimator=sum, ci=None, palette='plasma')
plt.xlabel('Teams')
plt.ylabel('Total Salary')
```

Data story : 
- In this above details we can stand out a conclusion, that employees under <b>Position 'C'</b> have spending more salary .ie,466377332.0.
- In case of teams,<b>Cleveland Cavaliers</b> have the highest total salary among the listed teams.ie,106,988,689.

<h5><b>5.Find if there is any correlation between age and salary , represent it visually.</b></h5>

```python
df2=df[['Age','Salary']]
df2
```

```python
corr = df2.corr()
print(corr)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix between Age and Salary")
plt.show()
```

Data story :

- Here "age" to "age" shows a positive correlation.
- "Age" to "Salary" indicates a positive but relatively weak relationship.
- "Salary" to "Age" indicates a  symmetric nature of correlation.
- Here "Salary" to "Salary" shows a positive correlation.
