```python
#pass rate
pass_rate=60
df['math pass_rate']=np.where(df['math score']<pass_rate,'Fail','Pass')
df
```

```python
df['math pass_rate'].value_counts()
```

```python
df['math score'].astype(np.int32)
```

```python
df['math score'].dtype
```

```python
#cross tabulation

math_table=pd.crosstab(index = df['gender'],columns=df['math pass_rate'])
math_table['PercentPass']=round(math_table['Pass']/(math_table['Pass'] + math_table['Fail'])*100,2)
math_table
```
